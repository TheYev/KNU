```python
import cv2
import mediapipe as mp

cap = cv2.VideoCapture(0)

wCam, hCam = 640, 480

cap.set(3,wCam)
cap.set(4,hCam)

mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

hands = mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5)

tipIds = [4, 8, 12, 16, 20]
```

```python
# Визначення функцій для пошуку пальців

def countFingers(image, hand_landmarks, handNo=0):
    
    if hand_landmarks:
        # Визначення всіх ключових точок для першої руки на відео
        landmarks = hand_landmarks[handNo].landmark
        
        # список пальців        
        fingers = []

        for lm_index in tipIds:
                # Визначення координати y початку і кінця пальця
                finger_tip_y = landmarks[lm_index].y 
                finger_bottom_y = landmarks[lm_index - 2].y

                # Процедура визначення відкритих і закритих конфігурацій
                if lm_index !=4:
                    if finger_tip_y < finger_bottom_y:
                        fingers.append(1)
                        print("FINGER with id ",lm_index," is Open")

                    if finger_tip_y > finger_bottom_y:
                        fingers.append(0)
                        print("FINGER with id ",lm_index," is Closed")

        totalFingers = fingers.count(1)

        # Display Text
        text = f'Fingers: {totalFingers}'

        cv2.putText(image, text, (50,50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0), 2)

# Визначення процедури малювання скелету руки 
def drawHandLanmarks(image, hand_landmarks):

    # Будуються сполучення (відрізки) між вузловими точками
    if hand_landmarks:
          for landmarks in hand_landmarks:
                mp_drawing.draw_landmarks(image, landmarks, mp_hands.HAND_CONNECTIONS)
```

```python
while True:
    success, image = cap.read()

    image = cv2.flip(image, 1)
    
    # Визначення ключових точок 
    results = hands.process(image)

    # Визначення позицій ключових точок
    hand_landmarks = results.multi_hand_landmarks

    # Малювання ключових точок
    drawHandLanmarks(image, hand_landmarks)

    # Визначення положення пальців        
    countFingers(image, hand_landmarks)

    cv2.imshow("Output", image)

    # Вихід з процедури після натистання клавіші Пробіл
    key = cv2.waitKey(1)
    if key == 32:
        break

cv2.destroyAllWindows()
```
